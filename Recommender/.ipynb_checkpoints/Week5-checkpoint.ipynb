```python
import graphlab
```

```python
song = graphlab.SFrame("song_data.gl")
```

```python
graphlab.canvas.set_target("ipynb")
```

```python
song.show()
```

```python
song.head()
```

```python
def artist_data(artist):
    name = song[song["artist"] == artist]
    return len(name["user_id"].unique())
```

## Counting unique users for singers

```python
print artist_data("Kanye West")
print artist_data('Foo Fighters')
print artist_data('Taylor Swift')
print artist_data('Lady GaGa')
```

## Find the most popular and least popular artist

```python
total_count = song.groupby(key_columns = "artist",  operations={'total_count': graphlab.aggregate.SUM("listen_count")})
total_count = total_count.sort("total_count")
```

```python
print total_count[0]
print total_count[-1]
```

## ML peronalized recommendation model

```python
train_data, test_data = song.random_split(.8, seed = 0)
```

```python
model = graphlab.item_similarity_recommender.create(train_data, user_id = "user_id",item_id = "song")
```

# Make recommendations for the 10000 users in the test data

```python
subset_test_users = test_data['user_id'].unique()[0:10000]
subset_test_users
```

# One recommended song for each of these test users

```python
recommendation = model.recommend(subset_test_users,k=1)
```

## Count how often each song is recommended

```python
data = recommendation.groupby(key_columns='song', operations={'count': graphlab.aggregate.COUNT()})
```

```python
data.sort('count', ascending=False)
```
